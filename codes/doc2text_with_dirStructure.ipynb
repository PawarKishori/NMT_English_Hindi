```python
## B4 running this cell first convert all doc to docx using:
#####    unoconv -d document --format=docx *.doc
#####     Run in system's python and not in conda python

## Conterting all docx to txt and adding underscore in filename and adding file id in filenames

import docx2txt

import os
path='aligned_doc_files/Aligned'
path1='aligned_doc_files/AlignedTXT'
drafts = os.listdir(path)
count=244
for eachdraft in drafts:
    print(count)
    for con in os.listdir(path+'/'+eachdraft):
#         print(os.listdir(path+'/'+eachdraft))
        if con !="Hindi":
            MY_TEXT = docx2txt.process(path+'/'+eachdraft+'/'+con)
            escriptname= "_".join(con.split('.')[:-1][0].split(" "))
            enewfile  = "E"+str(count)+'_'+escriptname+'.txt'
            print("eng: ", enewfile)

            with open(path1+'/'+enewfile, "w") as text_file:
                print(MY_TEXT, file=text_file)

        else:
            print(os.listdir(path+'/'+eachdraft+'/Hindi'))
            hn_file = os.listdir(path+'/'+eachdraft+'/Hindi')[0]

            MY_TEXT = docx2txt.process(path+'/'+eachdraft+'/Hindi/'+hn_file)
    
            hscriptname= "_".join(hn_file.split(".")[:-1][0].split(" "))
            
            hnewfile  = "H"+str(count)+'_'+hscriptname+'.txt'
            print("hin: ", hnewfile)            

        
            with open(path1+'/'+hnewfile, "w") as text_file:
                print(MY_TEXT, file=text_file)

    count=count+1
    print("")
    print("=============================================================")
#     break
# Run in system's python and not in conda python
#unoconv -d document --format=docx *.doc
```
