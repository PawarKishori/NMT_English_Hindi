#Input: Text file having Hindi text in paragraph style
#Output: Another text file with single hindi sentence per line
#Algorithm:
  ##Read the input file
  ##Split the text for new line delimiter in a list
  ##Create new text file.
  #For each paragraph, split for '.' and write to an output file

```python
import os;
```

```python
with open(os.path.join("/content/hindi_translated.txt"), "r",errors='ignore') as f:
  #hindi_text_encode = f.read().encode('utf-8').strip()
  hindi_text = f.read()
print('hindi_text',hindi_text)

#english_text = english_text.replace("\n", "")
#first_delimiter = bytes("\n", 'utf-8')
hindi_text_paras = hindi_text.split("\n")

#second_delimiter = bytes("|", 'utf-8')

for paras in hindi_text_paras:
  hindi_sentences = paras.split('।')
  for sentence in hindi_sentences:
    sentence = sentence + '।'
    print(sentence)
```
